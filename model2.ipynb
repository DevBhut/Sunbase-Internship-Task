```python
import pandas as pd
```

```python
df = pd.read_excel('customer_churn_large_dataset.xlsx')
```

```python
df.head()
```

```python
df.drop(['CustomerID', 'Name'], axis='columns', inplace=True)
```

```python
df.groupby('Churn')['Churn'].agg('count')
```

```python
df.head()
```

```python
df1 = pd.get_dummies(df.Location)
```

```python
df1.head()
```

```python
df.head()
```

```python
df2 = pd.concat([df, df1], axis = 'columns')
df2.head()
```

```python
df2.drop(['Location'], axis='columns', inplace=True)
df2.head()
```

```python
df3 = pd.get_dummies(df2.Gender)
df3.head()
```

```python
df4 = pd.concat([df2, df3], axis='columns')
df4.head()
```

```python
df4.drop(['Gender'], axis='columns', inplace=True)
df4.head()
```

```python
df4.shape
```

```python
y = df4.Churn
y.shape
```

```python
df5 = df4.drop(['Churn'], axis='columns')
```

```python
df5.head()
```

```python
df5.shape
```

```python
X = df5.copy()
X.head()
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, )
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix 
```

```python
print(accuracy_score(y_test, y_pred))
print(precision_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
